### Language Translator

We can easily translate any language using Google translator, but creating our translator will be a satisfying experience as a programmer. The googletrans API created by Google developers was used earlier to translate any text using Python. But now the new API from the Google developers known as **google_trans_new** is used for the same task.

We can use this API for every task that we can do with the google translator. we can easily install this library in our systems by using the **pip command**; `pip install google_trans_new`. To make a translator more interactive we will also create an interactive user interface by using the **streamlit library** in Python.
```python
from google_trans_new import google_translator
import streamlit as st
```

```python
translator = google_translator()
translator.translate("awesome", lang_tgt='fr') # translate to french language
```

```python
st.title("Language Translator")
text = st.text_input("Enter a text")
translate = translator.translate(text, lang_tgt='fr')
st.write(translate)
```

**run on terminal**

`streamlit run app.py`

Since we are using the streamlit library here, we need to run this code using the `streamlit run file_name.py` command in our terminal. After running the file a link will open in our default browser with an interactive user interface. In the text field, we can enter any text in any language, once we hit enter it will translate it to **French** as we have mentioned **fr** as our target language in the above code.
